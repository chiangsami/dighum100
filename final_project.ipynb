# Tokenizing 

```python
import string
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt' resource
nltk.download('punkt')
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    text_without_punctuation = text.translate(translator)
    return text_without_punctuation

all_tokens = []

for url in urls:
    print(f"Scraping {url}")
    print("-" * 50)
    response = requests.get(url)
    html_content = response.content

    soup = BeautifulSoup(html_content, 'html.parser')
    paragraphs = soup.find_all('p')

    if len(paragraphs) > 0:
        for paragraph in paragraphs:
            text = paragraph.text
            text_without_punctuation = remove_punctuation(text)
            text_lower = text_without_punctuation.lower()
            tokens = word_tokenize(text_lower)
            all_tokens.extend(tokens)
    else:
        print("No paragraphs found on the webpage.")
    
    print("\n")

# Print the final list of tokens
print(all_tokens)
```

```python
import string
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
import pandas as pd
from tabulate import tabulate

# Download the 'punkt' resource
nltk.download('punkt')

# Function to remove punctuation from a string
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    text_without_punctuation = text.translate(translator)
    return text_without_punctuation

# Selected words
climate_words = [
    'Weather', 'Atmosphere', 'Temperature', 'Precipitation', 'Wind',
    'Humidity', 'Seasons', 'Greenhouse effect', 'Global warming', 'Carbon dioxide',
    'Ozone layer', 'Renewable energy', 'Sustainability', 'Ecosystem', 'Natural resources',
    'Conservation', 'Emissions', 'Adaptation', 'Mitigation'
]

# Create an empty Counter to store the word counts
word_counts = Counter()

# Iterate through the URLs
for url in urls:
    print(f"Scraping {url}")
    print("-" * 50)
    
    # Send a GET request to the webpage and retrieve the HTML content
    response = requests.get(url)
    html_content = response.content

    # Create a Beautiful Soup object to parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all <p> tags on the webpage
    paragraphs = soup.find_all('p')

    # Check if any paragraphs exist
    if len(paragraphs) > 0:
        # Iterate through the paragraphs
        for paragraph in paragraphs:
            text = paragraph.text

            # Remove punctuation from the text
            text_without_punctuation = remove_punctuation(text)

            # Convert all words to lowercase
            text_lower = text_without_punctuation.lower()

            # Tokenize the text
            tokens = word_tokenize(text_lower)

            # Count the frequency of each word in the paragraph
            word_count = Counter(tokens)

            # Update the overall word counts
            word_counts.update(word_count)

    else:
        print("No paragraphs found on the webpage.")

# Filter the word count for selected words
climate_word_counts = {word: word_counts[word] for word in climate_words}

# Create a DataFrame from the word counts
df = pd.DataFrame.from_dict(climate_word_counts, orient='index', columns=['Frequency'])

# Add a column for Word Category
df['Category'] = 'Climate-related'

# Sort the DataFrame by frequency in descending order
df = df.sort_values('Frequency', ascending=False)

# Convert the DataFrame to a table string using tabulate
table = tabulate(df, headers='keys', tablefmt='fancy_grid')

# Display the table
print(table)
```
